```python
import asyncio
import time
# lets attempt Pulse Distance Modulation (PDM) vary the space between pulses to encode 0 or 1
async def pwm_for_dur(freq, duration, pwm_man):
    usec_conv = 1e6
    pwm_man.generate(int((1/freq)*usec_conv), 90)
    await asyncio.sleep(duration)
    pwm_man.stop()

async def blink_led():
    pwm_man = Pmod_IRTransceiver(base.PMODA, 0,1)
    send_bit = 1
    while True:
        await pwm_for_dur(38000, 0.05, pwm_man)
        await asyncio.sleep(0.05+send_bit*0.05)
        send_bit = (send_bit+1)%2

async def read_ir():
    lib = MicroblazeLibrary(base.PMODB, ['gpio'])
    gpio= lib.gpio_open(0)
    gpio.set_direction(lib.GPIO_IN)
    old = gpio.read()
    rising = time.time()
    print(f"Starting IR with value {old}")
    while True:
        val = gpio.read()
        if old<val:
            rising = time.time()
        elif old>val:
            print(f"pulse space: {time.time()-rising}")
        old=val
        await asyncio.sleep(0.01)

loop = asyncio.new_event_loop()
loop.create_task(blink_led())
loop.create_task(read_ir())
loop.run_forever()
loop.close()
```

```python
from pynq.lib.pmod.pmod_pwm import Pmod_PWM
import time

def tone_for_duration(freq, duration, pin):
    pwm_man = Pmod_PWM(base.PMODA, pin)
    usec_conv = 1e6
    pwm_man.generate(int((1/freq)*usec_conv), 90)
    time.sleep(duration)
    pwm_man.stop()
```
